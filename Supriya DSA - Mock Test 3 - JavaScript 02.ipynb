```python
#Implement a queue using an array in JavaScript. Include the necessary methods such as enqueue, dequeue, and isEmpty.

# Here's an implementation of a queue using an array in JavaScript:

class Queue {
  constructor() {
    this.items = [];
  }

  enqueue(element) {
    this.items.push(element);
  }

  dequeue() {
    if (this.isEmpty()) {
      return "Queue is empty.";
    }
    return this.items.shift();
  }

  isEmpty() {
    return this.items.length === 0;
  }

  size() {
    return this.items.length;
  }

  front() {
    if (this.isEmpty()) {
      return "Queue is empty.";
    }
    return this.items[0];
  }

  print() {
    console.log(this.items);
  }
}

// Example usage:
const queue = new Queue();
console.log(queue.isEmpty()); // true

queue.enqueue(10);
queue.enqueue(20);
queue.enqueue(30);

queue.print(); // [10, 20, 30]

console.log(queue.size()); // 3
console.log(queue.isEmpty()); // false
console.log(queue.front()); // 10

console.log(queue.dequeue()); // 10
console.log(queue.dequeue()); // 20

queue.print(); // [30]

"""In this implementation, the enqueue method adds an element to the end of the array, the dequeue method removes and 
   returns the first element from the array (using the shift method), and the isEmpty method checks if the queue is 
   empty by checking the length of the array. The front method returns the first element in the queue without removing 
   it, and the print method displays the contents of the array."""
```
